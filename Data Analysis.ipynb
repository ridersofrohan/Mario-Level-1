```python
'''
    0: Baseline
    1: Limited Actions
    2: Action History
    3: Function Approximation
'''
files = [
    './baseline/rewards.txt', 
    './pruned_actions/rewards.txt', 
    './action_history/rewards.txt',
    './func_approx/rewards.txt'
]
```

```python
data = []

for file in files:
    with open(file) as f:
        content = f.readlines()
        content = [eval(x.strip()) for x in content]
        data.append(content)
    f.close()
```

```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-dark-palette')
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
dataframes = []

for item in data:
    df = pd.DataFrame(item)
    dataframes.append(df)
```

```python
for i in range(len(dataframes)):
    df = dataframes[i]
    z = plt.figure()
    z.suptitle("Reward vs Epsiode", fontsize=16)
    plt.plot(df[0], df[1])
    plt.savefig('rewards-{}.png'.format(i+1))
    print("Rewards: Mean: {} \t STD: {}".format(df[0].mean(), df[0].std()))
    z = plt.figure()
    z.suptitle("Distance vs Epsiode", fontsize=16)
    plt.plot(df[0], df[2])
    plt.savefig('distance-{}.png'.format(i+1))
    print("Distance: Mean: {} \t STD: {}".format(df[1].mean(), df[1].std()))
```
